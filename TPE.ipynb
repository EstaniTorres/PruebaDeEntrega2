A continuación estudiaremos las hipótesis planteadas

Hipótesis 1: ProductRelatedDuration influye en Revenue (Evaluar si las sesiones que terminan en compra presentan mayor tiempo dedicado a páginas de productos). Utilizamos Mann-Whitney porque ProductRelatedDuration es continua, no normal y queremos comparar dos grupos independientes (compra vs. no compra).


```python
from scipy.stats import mannwhitneyu

compra = preprocessed_dataset[preprocessed_dataset["Revenue"]==1]["ProductRelated_Duration"]
no_compra = preprocessed_dataset[preprocessed_dataset["Revenue"]==0]["ProductRelated_Duration"]

stat, p = mannwhitneyu(compra, no_compra, alternative='greater')
print("Mann-Whitney U statistic:", stat)
print("p-value:", p)
```

Como el valor p < 0.05, existe evidencia estadísticamente significativa de que las sesiones con compra tienen una duración mayor en páginas de productos. Podemos concluir que H1 queda aceptada.

Hipótesis 2: Los usuarios recurrentes compran más (Analizaremos si existe relación entre el tipo de visitante y la compra final).
Para esta hipótesis utilizamos Chi-cuadrado porque tanto VisitorType como Revenue son variables categóricas y necesitamos evaluar si están asociadas.


```python
import scipy.stats as stats
import pandas as pd

tabla = pd.crosstab(preprocessed_dataset["VisitorType"], preprocessed_dataset["Revenue"])
chi2, p, dof, expected = stats.chi2_contingency(tabla)

print("Chi-cuadrado:", chi2)
print("p-value:", p)
```

El valor p es estramadamente menor que 0.05 por lo tanto existe una asociación estadísticamente significativa entre VisitorType y Revenue. Es decir, el comportamiento de compra cambia según el tipo de visitante. Concluimos que H2 queda aceptada.

Hipótesis 3: A mayor PageValue, mayor probabilidad de compra (verificamos si PageValue es significativamente mayor en sesiones que terminan en compra). Utilizamos Mann-Whitney porque PageValue es continua, altamente asimétrica, y comparamos su distribución entre sesiones con y sin compra.

```python
from scipy.stats import mannwhitneyu

compra = preprocessed_dataset[preprocessed_dataset["Revenue"]==1]["PageValues"]
no_compra = preprocessed_dataset[preprocessed_dataset["Revenue"]==0]["PageValues"]

stat, p = mannwhitneyu(compra, no_compra, alternative='greater')
print("p-value:", p)
```

Podemos afirmar que la hipótesis queda acpetada, es decir, que el valor de PageValues es significativamente mayor en sesiones donde hubo compra (cuanto mayor es el PageValue, mayor es la probabilidad de compra).

Hipótesis 4: Las sesiones sin compra navegan más páginas informativas (debemos determinar si los usuarios que no compran consumen más páginas de tipo informativo). Utilizamos Mann-Whitney porque Informational_Total es no normal y queremos comprobar si las sesiones sin compra muestran valores mayores que las sesiones con compra.

```python
from scipy.stats import mannwhitneyu

preprocessed_dataset["Informational_Total"] = (
    preprocessed_dataset["Administrative"] + preprocessed_dataset["Informational"]
)

compra_inf = preprocessed_dataset[preprocessed_dataset["Revenue"]==1]["Informational_Total"]
no_compra_inf = preprocessed_dataset[preprocessed_dataset["Revenue"]==0]["Informational_Total"]

stat, p = mannwhitneyu(no_compra_inf, compra_inf, alternative='greater')
print("p-value:", p)
```

No se observa ninguna evidencia de que las sesiones sin compra tengan más navegación informativa. Por lo tanto concluimos en que H4 queda rechazada

Hipótesis 5: Influencia combinada de ProductRelated_Duration, PageValues y ExitRates (vamos a evaluar si estas variables numéricas influyen conjuntamente en la probabilidad de compra). Usamos regresión logística porque Revenue es binaria y queremos evaluar el efecto conjunto de variables continuas sobre la probabilidad de compra.

```python
import statsmodels.api as sm

# Copiamos el dataframe de trabajo
df = preprocessed_dataset.copy()

# Aseguramos que Revenue sea numérica binaria (0/1)
df["Revenue"] = df["Revenue"].astype(int)

# Variables predictoras numéricas
X = df[["ProductRelated_Duration", "PageValues", "ExitRates"]]   # ajustar nombres si hace falta
y = df["Revenue"]

# Agregamos constante al modelo
X = sm.add_constant(X)

# Ajustamos el modelo logístico
modelo5 = sm.Logit(y, X).fit()

print(modelo5.summary())
```

Podemos afirmar que la hipótesis queda aceptada: las variables ProductRelated_Duration, PageValues y ExitRates influyen significativamente de manera conjunta sobre la probabilidad de compra. En particular, PageValues y ProductRelated_Duration incrementan la probabilidad de compra, mientras que ExitRates la reduce.

Para contribuir a la hipótesis, usamos tres boxplots (uno por variable numérica) porque cada predictor tiene una escala diferente y distribuciones muy distintas entre sí. Es la forma más clara de comparar la distribución de la variable entre sesiones con y sin compra, nos permite ver diferencias en mediana, dispersión y presencia de outliers para cada predictor, lo cual complementa visualmente la evidencia estadística obtenida en la regresión logística.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
sns.boxplot(data=df, x="Revenue", y="ProductRelated_Duration")
plt.title("ProductRelated_Duration vs Revenue")

plt.subplot(1,3,2)
sns.boxplot(data=df, x="Revenue", y="PageValues")
plt.title("PageValues vs Revenue")

plt.subplot(1,3,3)
sns.boxplot(data=df, x="Revenue", y="ExitRates")
plt.title("ExitRates vs Revenue")

plt.tight_layout()
plt.show()
```

Hipótesis 6: Influencia conjunta de BounceRates y ExitRates en la probabilidad de compra (Analizaremos las tasas de abandono del sitio (BounceRates y ExitRates) influyen negativamente en la probabilidad de compra). Usamos regresión logística multivariada porque Revenue es una variable binaria y queremos evaluar el efecto conjunto de dos predictores numéricos continuos sobre la probabilidad de compra.

```python
import statsmodels.api as sm

X = preprocessed_dataset[["BounceRates", "ExitRates"]]
y = preprocessed_dataset["Revenue"].astype(int)

X = sm.add_constant(X)

modelo6 = sm.Logit(y, X).fit()
print(modelo6.summary())
```

El modelo multivariado muestra que solo ExitRates influye significativamente en la probabilidad de compra, mientras que BounceRates no presenta efecto estadístico. En consecuencia, aceptamos H6 parcialmente.
